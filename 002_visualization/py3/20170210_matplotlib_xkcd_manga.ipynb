```python
# https://jp.mathworks.com/help/matlab/ref/linspace.html?requestedDomain=jp.mathworks.com
# -pi 〜 pi の間の等間隔な点を 200個含むベクトルを返却する

def plot_sincon(mode):
    plt.xkcd(mode)
    x = np.linspace(-np.pi, np.pi, 200)
    
    plt.plot(x, np.sin(x), label="y = sinx")
    plt.plot(x, np.cos(x), label="y = cosx")
    plt.annotate(
        'Intersection of \n sinx and cosx',
    xy=(np.pi/4, np.sin(np.pi/4.0)),
    arrowprops=dict(arrowstyle='->'), 
    xytext=(0, -0.5)
    )
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.xlim(-np.pi, np.pi)
    plt.ylim(-1, 1)
    plt.title("normal")
    plt.legend(loc = "upper left")
    plt.show()

plot_sincon(False)
plot_sincon(True)

```
